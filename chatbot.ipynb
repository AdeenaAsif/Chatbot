```python
import random
import re

class ChatBot:
    neg_res = ['no', 'bye', 'sorry', 'impossible', 'nope', 'no thanks']
    exit_res = ['bye', 'goodbye', 'exit', 'quit', 'best wishes', 'farewell','That`s all','thanks']

    def __init__(self):
        self.support_res = {
            'ord_track': r'\b(track)\b',
            'return_prod': r'\b(return| return policy)\b',
            'complements': r'\b(discounts|deals|sales|sale|event)\b',
            'size_guide':r'\b(size|fit|loose|unfit)\b',
            'general_QA':r'\b(name)\b'
            
        }

    def greets(self):
        self.name = input("Greetings from ChicCraze! May I know your pretty name Please?\n")
        helper = input(f"Hi {self.name}. How can I make your day more stylish?\n")
        if helper in self.neg_res:
            print(f"Alright {self.name}. If you change your mind, I'll be here. Have a fantastic day!")
        else:
            return self.chat()

    def quit(self, response):
        for reply in self.exit_res:
            if reply in response:
                print("Thankyou for chatting with ChicCraze.If style questions pop up, I'm just a message away. Wishing you a fabulous day!")
                
                return True
        return False
    def chat(self):
        response = input("Please feel free to ask your queries:\n ").lower()
        while not self.quit(response):
            response = input(self.match_chat(response)+'\n')

    def match_chat(self, response):
        for x, y in self.support_res.items():
            pattern = re.search(y, response)
            if pattern and x == 'ord_track':
                return self.ord_track()
            elif pattern and x == 'return_prod':
                return self.return_prod()
            elif pattern and x == 'complements':
                return self.compliments()
            elif pattern and x =='size_guide':
                return self.size_guide()
            elif pattern and x == 'general_QA':
                return self.general_QA()
        return self.no_match()
            

    def ord_track(self):
        answers = ["To track your order, simply visit our website's 'Track Order' page and enter your order number.",
                   "Get instant order updates by texting your order number to our dedicated tracking number.",
                   "Track your package effortlessly by scanning the QR code on your order confirmation."]
        return random.choice(answers)

    def compliments(self):
        answers = ["Absolutely! Keep an eye on our website's 'Promotions' page for announcements about upcoming sales and events.",
                   "Certainly! Subscribe to our newsletter to be the first to know about upcoming sales, events, and special promotions.",
                   "Yes, we're planning something special! Join our loyalty program for early access and exclusive offers.",
                   "I'm sorry, but as of now, I don't have information regarding any upcoming sales.",
                   "I don't have access to details about future events or sales at this moment."]
        return random.choice(answers)

    def return_prod(self):
        answers = ["Return items within 30 days for a full refund. Visit our 'Returns' page for instructions.",
                   "Return online purchases within 30 days. Find details in our 'Help' section."]
        return random.choice(answers)
    def size_guide(self):
        
        answers= ["Our Size & Fit Guide is designed to help you find the perfect fit. Check it out on our website!",
             "We understand the importance of a perfect fit. Explore our Size & Fit Guide for detailed information.",
             "Concerned about the fit? Visit our website's Size & Fit Guide for comprehensive sizing information.",
             "Finding the right size is crucial. Head to our Size & Fit Guide on the website for assistance.",
             "Worried about the fit? Our Size & Fit Guide is here to assist you in choosing the right size."]
        return random.choice(answers)

    def general_QA(self):
        answers= ["My name is ChicCraze, your style assistant!",
            "I'm your friendly ChicCraze assistant, here to make your style journey delightful."]
        return random.choice(answers)
    def no_match(self):
        answers=["I'm not quite getting it. Can you try rephrasing your question?",
                "I'm having a moment of confusion. Can you walk me through that again?"
                "I'm not entirely clear on that. Can you provide additional context?"]
        return random.choice (answers)
chat_bot = ChatBot()
chat_bot.greets()
```
